```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
# Read file from this path
path='/content/drive/MyDrive/Colab Notebooks/Swimming Python Project/800Free.csv'
event=pd.read_csv(path, encoding= 'unicode_escape')
```

```python
# Presentation of the dataset
event
```

```python
# Create new columns in the dataset

s=np.random.randint(1,101,event.shape[1])

event['Time'] = s
event['Name'] = s
event['Club'] = s
event['Date'] = s
event['Place'] = s
```

```python
# Split the dataset and assign the values ​​in the corresponding columns

i=0
for x in event['Time']:              #Change the name if necessary
  lines = x.split()
  if i==0 or i==6 or i==8 or i==9:   #Change the indexs if necessary for names with more than 3 words
    event['Time'][i] = lines[0]
    event['Name'][i] = lines[1]+ ' ' + lines[2] + ' ' + lines[3] + ' ' + lines[4]
    event['Date'][i] = lines[5]
    event['Date'][i] = lines[6]
    event['Place'][i] = lines[7:]
  else:
    event['Time'][i] = lines[0]
    event['Name'][i] = lines[1]+ ' ' + lines[2] + ' ' + lines[3]
    event['Club'][i] = lines[4]
    event['Date'][i] = lines[5]
    event['Place'][i] = lines[6:]
  i=i+1
```

```python
# Correct specific values ​​in a given column

col = ' '
idx = 1
value = ' '

event[col][idx] = value
```

```python
# Imput new rows in the dataset

number = int(input('Number of Times: '))
i=0
while(i<number):
  print(i+1)
  tempo = input('Time: ')
  nome = input('Name: ')
  clube = input('Club: ')
  data = input('Date: ')
  event.loc[len(event.index)] = [tempo, nome, clube, data] 
```

```python
# Drop columns from dataset
event=event.drop(columns=['Place'])
```
