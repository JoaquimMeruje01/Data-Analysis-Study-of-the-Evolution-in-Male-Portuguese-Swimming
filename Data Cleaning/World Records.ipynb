```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Read file from this path
path='/content/drive/MyDrive/Colab Notebooks/Swimming Python Project/800Free.csv'
event=pd.read_csv(path, encoding= 'unicode_escape')
```

```python
# Presentation of the dataset
event
```

```python
# Delete the first asterisks from the 'Time' column

ch = '*'
i = 0
while(i<event.shape[0]):
  event['Time'][i]=event[1:]   
  i=i+1
```

```python
# Delete the last asterisks from the 'Time' column

ch = '*'
i = 0
while(i<event.shape[0]):
  event['Time'][i]=event[0:-2]   
  i=i+1
```

```python
# Delete strange characters from the 'Time' column

i = 0
while(i<event.shape[0]):
  for x in event['Time'][i]:
    if x.find("'") != -1:
      event['Time'][i] = event['Time'][i].replace("'",'.')
    if x.find('"') != -1:
      event['Time'][i] = event['Time'][i].replace('"',':')  
  i=i+1
```

```python
# Drop columns from dataset

event=event.drop(columns=['Place'])
```

```python
# Correct specific values ​​in a given column

col = ' '
idx = 1
value = ' '

event[col][idx] = value
```

```python
# Imput new rows in the dataset

number = int(input('Number of Times: '))
i=0
while(i<number):
  print(i+1)
  tempo = input('Time: ')
  nome = input('Name: ')
  data = input('Date: ')
  event.loc[len(event.index)] = [tempo, nome, data] 
  i=i+1
```
